```python
def tonelli_shanks(a, p):
    """Алгоритм Тонелли-Шенкса для нахождения квадратного корня по модулю простого p"""
    assert pow(a, (p - 1) // 2, p) == 1, "a должно быть квадратичным вычетом по модулю p"
    
    # Если a = 0, то корень 0
    if a == 0:
        return 0

    # Если p ≡ 3 (mod 4), можно просто вычислить (a^(p+1)/4) % p
    if p % 4 == 3:
        return pow(a, (p + 1) // 4, p)

    # Инициализация переменных
    s, q = 0, p - 1
    while q % 2 == 0:
        s += 1
        q //= 2

    # Ищем неполный квадрат (z), который не является квадратным вычетом
    z = 2
    while pow(z, (p - 1) // 2, p) == 1:
        z += 1

    # Инициализация переменных
    m = s
    c = pow(z, q, p)
    t = pow(a, q, p)
    r = pow(a, (q + 1) // 2, p)

    while t != 0 and t != 1:
        # Находим наименьший i такой, что t^(2^i) ≡ 1 (mod p)
        t2i = t
        i = 0
        while t2i != 1:
            t2i = pow(t2i, 2, p)
            i += 1

        # Обновляем m, c, t, r
        b = pow(c, 2 ** (m - i - 1), p)
        m = i
        c = pow(b, 2, p)
        t = (t * c) % p
        r = (r * b) % p

    if t == 0:
        return 0
    return r


root = tonelli_shanks(a, p)
print(root)
```
